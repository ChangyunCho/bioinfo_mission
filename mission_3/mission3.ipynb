```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import seaborn as sns
from IPython.display import Image
from scipy.stats import entropy
import re
```

## Mirlet7g

```python
pileup = pd.read_csv('/data/project/joebrother/lecture/2021_bioinfo/binfo1/mission_2/binfo1-datapack1/CLIP-let7g-gene.pileup', sep = '\t', names=['chrom','pos','_ref','count','basereads','quals'])
```

```python
toremove = re.compile('[<>$*#^]')
pileup['matches'] = pileup['basereads'].apply(lambda x: toremove.sub('', x))
```

```python
pileup[['chrom', 'pos', 'matches']]
```

```python
pileup[pileup['pos'] == 106056094].iloc[0]['matches']
```

```python
pileup['match_count'] = pileup['matches'].apply(lambda x: len(x))
```

```python
pileup['A'] = 0
pileup['T'] = 0
pileup['G'] = 0
pileup['C'] = 0

for i in range(len(pileup)):
    pileup['A'][i] = pileup['matches'][i].upper().count('A')
    pileup['T'][i] = pileup['matches'][i].upper().count('T')
    pileup['G'][i] = pileup['matches'][i].upper().count('G')
    pileup['C'][i] = pileup['matches'][i].upper().count('C')
```

```python
pileup['entropy'] = entropy(pileup[['A', 'T', 'G', 'C']].T, base=2)
pileup
```

```python
pileup['start'] = pileup['pos'].apply(lambda x: x-1)
pileup_screened = pileup[['chrom', 'start', 'pos', 'entropy']].dropna()
pileup_screened.to_csv('let7g_entropy.bedgraph', sep='\t', index=None, header=None)
```
